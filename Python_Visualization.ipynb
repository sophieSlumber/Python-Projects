```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
df = sns.load_dataset('diamonds')
```

```python
df.info()
df.shape
```

```python
df.sample(5)
```

## Histograms and KDEs

```python
#Plot a histogram of diamond prices with bins equal to 20
axx = sns.histplot(df['price'], bins = 20)
```

```python
ax = sns.histplot(np.log(df['price']),  bins = 20 );
```

```python
# Plot a kernel density estimate plot of diamond prices
sns.kdeplot(data=df, x=np.log(df['price']));
```

```python
#Plot a histogram of diamond prices with bins equal to 1000
x = np.arange(0,20000,1000)

fig, ax = plt.subplots()
ax.set_title('Count of diamonds by price')
plt.xticks(x)
plt.xticks(rotation=90)#rotate the axis labels

y = sns.histplot(df.price, element='bars', bins=x, stat='count', legend=True)

#add index labels
for i in y.containers:
    y.bar_label(i,fontsize=6,color='m')
```

## Histograms and distribution types

```python
# Plot a histogram of 100000 normal random variables, split across 40 bins
sns.histplot(np.random.normal(size = 100000),  bins = 40 ); 
plt.xlabel('Value')
plt.title("normal random distribution")
```

```python
# Plot a right skewed histogram of 100000 normal random variables, 
# split across 40 bins, (using the beta distribution)

sns.histplot(np.random.beta(a=0.3, b = 1,size = 100000),  bins = 40 );
```

```python
# Plot a left skewed histogram 

sns.histplot(np.random.beta(a=1 , b = 0.3,size = 100000),  bins = 40 ); 
```

## Bar charts

```python
df['color'].value_counts().sort_values(ascending=True)
```

```python
plt.figure(figsize = (8,5))
plt.rcParams['font.size'] = 10
plot = sns.countplot(x='color',data=df, order = np.sort(df['color'].unique()))

for p in plot.containers:
    plot.bar_label(p,fontsize=10,color='k')
```

## Scatterplot

## Heatmaps

```python
flights_long = sns.load_dataset('flights')
```

```python
flights_long.sample(10)
```

```python
flights = flights_long.pivot(index = 'month',
    columns = 'year', values = 'passengers')
plt.figure(figsize = (7,7))
plt.rcParams['font.size'] = 8
sns.heatmap(flights, annot=True, fmt='d')
```

## Bokeh library

```python
from bokeh.plotting import figure, show

p = figure(width=400, height=400)
p.hbar(y=[1, 2, 3], height=0.5, left=0,
       right=[1.2, 2.5, 3.7], color="navy")

show(p)
```

```python
from bokeh.plotting import figure

# set up a figure with a toolbar
p = figure(
    height=300,
    tools=("box_zoom", "wheel_zoom", "save", "reset"),  # 🔁 swap out tools from the list above
)

p.line(x=list(range(10)), y=list(range(10)))
show(p)
```
