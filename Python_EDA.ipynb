### Covariance and Correlations

```python
diamonds.cov(numeric_only=True)
```

```python
diamonds.corr(numeric_only=True)
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(diamonds.corr(numeric_only=True),annot=True)
```

```python
diamonds[['carat','price','x']].corr()
```

```python
#create a heatmap of correlations above

sns.heatmap(diamonds[['carat','price','x']].corr(), annot=True)
plt.show()
```

```python
# A scatterplot between price and carat

sns.set_style(style='whitegrid')
plt.figure(figsize=(8,5))
sns.scatterplot(data=diamonds.sample(900), x='carat',y='price')
```

```python
# Phi-k correlations

import phik
X = diamonds.phik_matrix()
X
```

```python
sns.heatmap(X,annot=True)
```

```python
from phik import report
phik.report.correlation_report(diamonds)
```

### Quantiles Distribution

```python
diamonds.quantile(0.30,numeric_only=True)
```

```python
diamonds.quantile([0.1,0.3,0.5,0.7],numeric_only=True)
```

```python
# create a quantile plot

import statsmodels.api as sm

fig = sm.qqplot(diamonds['carat'])
plt.show()
```

### Pivot Tables

```python
diamonds.head(10)
```

```python
diamonds.pivot_table(index = ['cut','color'],values = 'price')
```

```python
diamonds.pivot_table(index = ['cut','color'],values = 'price',
                    aggfunc = ['mean','count','median','sum'])
```

```python
diamonds.pivot_table(index = ['clarity', 'cut'],
              columns = ['color'],
              values = ['depth', 'price', 'x'],
              aggfunc = {'depth': np.mean,
                        'price': [min, max, np.median],
                        'x': np.median}
              )
```

### Crosstab

```python
pd.crosstab(diamonds.cut,diamonds.color)
```

```python
# with margins
pd.crosstab(diamonds.cut,diamonds.color,margins=True)
```

```python
# with margins and normalized

pd.crosstab(diamonds.cut,diamonds.color,margins=True,normalize=True)
```

### Groupby

```python
diamonds.groupby(['cut','color']).agg({'price':'mean','x':'count',
                                      'carat':'mean'}).rename(columns={'x':'count'})
```

```python
diamonds.groupby(['color']).agg('mean',numeric_only=True)
```

```python
diamonds.groupby('cut').agg({'price':'sum',
                            'clarity':'count'}).rename(columns={'price':'total_price','clarity':'diamonds_count'})
```
